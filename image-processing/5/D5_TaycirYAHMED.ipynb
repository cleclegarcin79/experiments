#### Algorithm implementation 

```python
def midway_eq(u,v):
    
    u_f = u.flatten()
    v_f = v.flatten()
    u_index = np.argsort(u_f)
    v_index = np.argsort(v_f)
    u_sorted = np.sort(u_f)
    v_sorted = np.sort(v_f)
    
    u_midway =  np.zeros(u_f.shape)
    v_midway =  np.zeros(v_f.shape)
    
    u_midway[u_index] = (u_sorted + v_sorted)/2
    v_midway[v_index] = (u_sorted + v_sorted)/2
    
    return u_midway.reshape(u.shape), v_midway.reshape(v.shape)
        
```

```python
u_n , v_n = midway_eq(img, img2)
```

## Zoom with implicit TCD

### Zoom with zero padding 

```python
img_gray = cv2.imread('images/house.png',0)
```

```python
k = 5

u_chap = fftpack.fft2(img_gray)
u_chap = fftpack.fftshift(u_chap)

N = u_chap.shape[0] 

u_padd = np.zeros((N*k, N*k), dtype='complex64')

nb = u_padd.shape[0]
na = u_chap.shape[0]
lower = (nb) // 2 - (na // 2)
upper = (nb // 2) + (na // 2)
u_padd[lower:upper, lower:upper] = u_chap


u_padd = fftpack.ifftshift(u_padd)
u_padd = np.real(fftpack.ifft2(u_padd))

print "Image size =" , u_padd.shape
imgplot = plt.imshow(u_padd, cmap='gray')
```

### Zoom with cosinus 

```python
k = 5

u_chap = fftpack.dct(fftpack.dct(img_gray.T).T)


u_chap = fftpack.fftshift(u_chap)

N = u_chap.shape[0] 

u_padd = np.zeros((N*k, N*k), dtype='complex64')

nb = u_padd.shape[0]
na = u_chap.shape[0]

lower = (nb) // 2 - (na // 2)
upper = (nb // 2) + (na // 2)
u_padd[lower:upper, lower:upper] = u_chap


u_padd = fftpack.ifftshift(u_padd)

u_padd = np.real(fftpack.idct(fftpack.idct(u_padd.T).T))

print "Image size =" , u_padd.shape
imgplot = plt.imshow(u_padd, cmap='gray')
```
